```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("base_total.xlsx", engine="openpyxl")
df = df.convert_dtypes()
```

```python
filtered = df[df["Corte"].dt.normalize() == pd.Timestamp("2025-01-01") ].copy()
```

```python
filtered.rename(columns={'Freceuncia_Pago':'frecuencia_pago'}, inplace= True)
filtered = filtered[filtered['frecuencia_pago'] == 'MENSUAL']

filtered.head()
```

Quitar medios de pago de productos diferentes

```python
filtered = filtered[(~ filtered["MedioPago"].isin(["CANCELACIONES","SINIESTROS","COSTUMER ENGAGEMENT"])) & filtered["MedioPago"].notna()]
```

```python
filtered["MedioPago"].value_counts(dropna=False)
```

```python
filtered["MedioPago"] = np.where(filtered["MedioPago"] == "PAGO DIRECTO", "MANUAL", "AUTOMATICO")
```

```python
filtered["MedioPago"].value_counts(dropna=False)
```

```python
filtered.columns
```
